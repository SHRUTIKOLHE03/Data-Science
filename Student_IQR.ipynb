```python
for col in nc:
    sns.boxplot(x=df[col])
    plt.title("Box PLot")
    plt.show()
 
```

```python
for col in nc:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lb = Q1 - 1.5*IQR
    ub = Q3 + 1.5*IQR
    df[col] = np.where((df[col]<lb) | (df[col]>ub), df[col].median(), df[col])

print(df)
```
