```python
#finding outlier
q1=df['sepal_length'].quantile(0.25)
q3=df['sepal_length'].quantile(0.75)
q1
q3
iqr=q3-q1

lb=q1-1.5*iqr
ub=q3+1.5*iqr

lb
ub
```

```python
df.fillna(df[df['sepal_length']<lb & df['sepal_length']>ub],df['sepal_length'].median(), inplace =True)
```
